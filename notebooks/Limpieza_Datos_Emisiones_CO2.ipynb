```python
import pandas as pd

# 1. Cargar el dataset
df = pd.read_csv('GCB2022v27_MtCO2_flat.csv')

# 2. Filtrar para quedarnos solo con datos modernos (desde 1900)
# Esto reduce el ruido de los años donde casi todo es 0
df_clean = df[df['Year'] >= 1900].copy()

# 3. Rellenar valores nulos (NaN) con 0
# En este dataset, si no hay dato es porque no hubo emisión de ese tipo
df_clean = df_clean.fillna(0)

# 4. Eliminar la columna 'Other' si no la vamos a usar para simplificar
if 'Other' in df_clean.columns:
    df_clean = df_clean.drop(columns=['Other'])

# 5. Guardar el archivo limpio para SQL
df_clean.to_csv('CO2_Emissions_Clean.csv', index=False)

print("¡Archivo limpio generado con éxito! Descárgalo de la carpeta de la izquierda.")
```
